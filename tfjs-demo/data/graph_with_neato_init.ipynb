```python
fn_groups = [
#     [
#         'graphs/bipartite-graph-5-5.json',
#         'neato_sfdp_layouts_json/bipartite_neato.json',
#         'neato_sfdp_layouts_json/bipartite_sfdp.json'
#     ],
#     [
#         'graphs/block-5-5.json',
#         'neato_sfdp_layouts_json/block_neato.json',
#         'neato_sfdp_layouts_json/block_sfdp.json'
#     ],
#     [
#         'graphs/cube.json',
#         'neato_sfdp_layouts_json/cube_neato.json',
#         'neato_sfdp_layouts_json/cube_sfdp.json'
#     ],
#     [
#         'graphs/cycle-10.json',
#         'neato_sfdp_layouts_json/cycle_neato.json',
#         'neato_sfdp_layouts_json/cycle_sfdp.json'
#     ],
#     [
#         'graphs/dodecahedron.json',
#         'neato_sfdp_layouts_json/dodecahedron_neato.json',
#         'neato_sfdp_layouts_json/dodecahedron_sfdp.json'
#     ],
#     [
#         'graphs/grid-5-5.json',
#         'neato_sfdp_layouts_json/grid_neato.json',
#         'neato_sfdp_layouts_json/grid_sfdp.json'
#     ],
#     [
#         'graphs/input-9.json',
#         'neato_sfdp_layouts_json/spx_teaser_neato.json',
#         'neato_sfdp_layouts_json/spx_teaser_sfdp.json'
#     ],
    [
        'graphs/nonsymmetric.json',
        'neato_sfdp_layouts_json/nonsymmetric_neato.json',
        'neato_sfdp_layouts_json/nonsymmetric_sfdp.json'
    ],
#     [
#         'graphs/tree-2-3.json',
#         'neato_sfdp_layouts_json/tree_neato.json',
#         'neato_sfdp_layouts_json/tree_sfdp.json'
#     ],
#     [
#         'graphs/k-20.json',
#         'neato_sfdp_layouts_json/complete_neato.json',
#         'neato_sfdp_layouts_json/complete_sfdp.json'
#     ],
]

for fn, fn_neato, fn_sfdp in fn_groups:
    print(fn)
    
    with open(fn) as f, \
        open(fn_neato) as f_neato, \
        open(fn_sfdp) as f_sfdp:
        
        graph = json.load(f)
        graph_neato = json.load(f_neato)
        graph_sfdp = json.load(f_sfdp)
        
    graph['nodes'] = [{'index': d['index'], 'id': d['id']} for d in graph_neato['nodes']]
    graph['edges'] = graph_neato['edges']
    
    init_neato = [[d['x'],d['y']] for d in graph_neato['nodes']]
    init_sfdp = [[d['x'],d['y']] for d in graph_sfdp['nodes']]

    graph['initPosition_neato'] = init_neato
    graph['initPosition_sfdp'] = init_sfdp
        
    fn_out = fn.split('/')[-1]
    with open(fn_out, 'w') as fout:
        json.dump(graph, fout, indent=2)
```
